```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('2022_advanced_batting.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df = df.drop(columns=['-9999'], axis=1)
```

```python
df.info()
```

```python
df.columns = df.columns.str.lower()
```

```python
df.info()
```

```python
df = df.drop(columns=['rk'], axis=1)
```

```python
df.info()
```

```python
df['hr%'] = df['hr%'].astype(str).str.rstrip('%').astype('float') / 10.0
```

```python
df.head()
```

```python
df['so%'] = df['so%'].astype(str).str.rstrip('%').astype('float') / 100
```

```python
df['bb%'] = df['bb%'].astype(str).str.rstrip('%').astype('float') / 10.0
df['hardh%'] = df['hardh%'].astype(str).str.rstrip('%').astype('float') / 10.0
df['ld%'] = df['ld%'].astype(str).str.rstrip('%').astype('float') / 100
df['gb%'] = df['gb%'].astype(str).str.rstrip('%').astype('float') / 100
df['fb%'] = df['fb%'].astype(str).str.rstrip('%').astype('float') / 100
df['pull%'] = df['pull%'].astype(str).str.rstrip('%').astype('float') / 100
df['cent%'] = df['cent%'].astype(str).str.rstrip('%').astype('float') / 100
df['oppo%'] = df['oppo%'].astype(str).str.rstrip('%').astype('float') / 100
df['cwpa'] = df['cwpa'].astype(str).str.rstrip('%').astype('float')
df['rs%'] = df['rs%'].astype(str).str.rstrip('%').astype('float') / 100.0
df['sb%'] = df['sb%'].astype(str).str.rstrip('%').astype('float') / 100.0
df['xbt%'] = df['xbt%'].astype(str).str.rstrip('%').astype('float') / 100.0
```

```python
df.info()
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df = df.dropna(subset=['roba'])
```

```python
df.isna().sum()
```

```python
df['babip'] = df['babip'].fillna(.000)
```

```python
df.head(50)
```

```python
df = df.dropna(subset=['iso'])
```

```python
df.isna().sum()
```

```python
df = df.dropna(subset=['ev'])
```

```python
df.isna().sum()
```

```python
df['gb/fb'] = df['gb/fb'].fillna(0.00)
```

```python
df['rs%'] = df['rs%'].fillna(0.00)
```

```python
df['sb%'] = df['sb%'].fillna(0.00)
```

```python
df['xbt%'] = df['xbt%'].fillna(0.00)
```

```python
df.isna().sum()
```

```python
hr_df = df.loc[df['pa'] >= 50]
```

```python
hr_df.head()
```

```python
hr_df.info()
```

```python
plt.figure(figsize=(16,8))
ax = sns.scatterplot(data=hr_df, x='so%', y='hr%')
plt.title('Home Run % and Strikeout % For Major League Batters (Min. 50 PA)', fontsize=14)
plt.xlabel('Strikeout %', fontsize=12)
plt.ylabel('HR %', fontsize=12)
```
